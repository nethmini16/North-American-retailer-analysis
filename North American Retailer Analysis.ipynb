Removed 52 null values in both the columns ‘Customer Province’ and ‘Item Stock Status’ as it was only a very small amount. After removing these null values, the number of rows was reduced from 10,000 to 9,948 rows.

The same rationale was applied to warehouse-related columns and company-related columns as well. #### Checking for any inconsistencies in Ware house name and Ware house Region Name

Inconsistencies between columns that should ideally be consistent with each other were checked such as the ‘Warehouse Name’ and ‘Warehouse Region Name’

```python
warehouse_list = list(zip(df_new["WAREHOUSE_NAME"].values, df_new["WAREHOUSE_REGION_NAME"].values))
warehouse_list
```

```python
inconsistent = []
warehouse_dict = {}

for name, region in warehouse_list:
    
  if name not in warehouse_dict:
    warehouse_dict[name] = region
  else:
    if warehouse_dict[name] != region:
      inconsistent.append(name)
```

```python
#Checking if there are any inconsistent records in the inconsistent list
inconsistent
```


#### Checking for any inconsistencies in Item Number and Item Department

```python
item_list = list(zip(df_new["ITEM_NUMBER"].values, df_new["ITEM_DEPARTMENT"].values))
item_list
```

```python
inconsistent = []
item_dict = {}

for item, dep in item_list:
    
  if item not in item_dict:
    item_dict[item] = dep
  else:
    if item_dict[item]  != dep:
      inconsistent.append(item)
```

```python
#Checking if there are any inconsistent records in the inconsistent list
inconsistent
```

#### Creating a new column for Revenue, Cost and Profit

New columns for Revenue, Cost and Profit were created. Revenue was considered as ‘Sales Amount’ multiplied by the ‘Order Quantity’ and Cost was calculated as ‘Cost Price’ multiplied by the ‘Order Quantity’. Thereafter, Profit was calculated by deducting the ‘Cost’ from the ‘Revenue’

```python
df_new["REVENUE"] = df_new["SALES_AMOUNT"] * df_new["ORDER_QUANTITY"]
```

```python
df_new["COST"] = df_new["COST_PRICE"] * df_new["ORDER_QUANTITY"]
```

```python
df_new["PROFIT"] = df_new["REVENUE"] - df_new["COST"]
```
